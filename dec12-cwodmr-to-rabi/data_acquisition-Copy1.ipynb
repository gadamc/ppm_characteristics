```python

def plot_trace(trace, N_cycles=1, clock_rate=1, label = None):
    plt.plot(np.arange(len(trace)) / clock_rate, trace * clock_rate / N_cycles, label = label)
    
def get_nearest_frequency(scan_data, freq):
    frequencies = np.array([x[0] for x  in scan_data ])
    idx = (np.abs(frequencies - freq)).argmin()
    return idx, frequencies[idx]

def plot_frequency(scan_data,  experiment, freq = 2870e6):
    idx, nearest_freq = get_nearest_frequency(scan_data, freq)
    scan = scan_data[idx]
    plot_trace(scan[1], experiment.N_cycles, clock_rate = 1./experiment.pulser.clock_period, label = scan[0])
    return nearest_freq
```

```python
#pulled this up from below --
def contrast_calculator(data_buffer, experiment):
    trace = qt3utils.experiments.common.aggregate_sum(data_buffer, experiment)
    rf_on = trace[len(trace)//8:len(trace)//2]
    rf_off = trace[5*len(trace)//8:]
    return np.sum(rf_on)/np.sum(rf_off)
```

```python
def counts_to_rate(trace, experiment):
    return trace/experiment.N_cycles / experiment.pulser.clock_period

def threshold_suppress(trace, experiment):
    trace[trace < 1e4] = 0
    return trace

def signal_extract(trace, experiment):
    trace = threshold_suppress(counts_to_rate(trace, experiment), experiment)
    background = trace[:len(trace)//2]
    signal = trace[len(trace)//2:]
    return np.sum(signal)/np.sum(background)

def podmr_contrast(data_buffer, experiment):
    trace = qt3utils.experiments.common.aggregate_sum(data_buffer, experiment)
    return signal_extract(trace, experiment)
```

```python
def counts_thresh(t, e):
    return threshold_suppress(counts_to_rate(t,e),e)

def raw_to_processed(scan_data, f):
    return [[x[0], f(x[1], podmr_exp)] for x in scan_data]
```

```python
def podmr_contrast(data_buffer, experiment):
    trace = qt3utils.experiments.common.aggregate_sum(data_buffer, experiment)
    trace = trace/experiment.N_cycles / experiment.pulser.clock_period
    trace[trace < 1e4] = 0
    background = trace[:len(trace)//2]
    signal = trace[len(trace)//2:]
    return np.sum(signal)/np.sum(background)
```
